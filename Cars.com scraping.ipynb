Imports

```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

HTTP Request

Storing website in variable

```python
website = 'https://www.cars.com/shopping/results/?list_price_max=&maximum_distance=30&stock_type=used&zip=75010'
```

Get request

```python
response = requests.get(website)
```

Status Code

```python
response.status_code
```

Soup Object

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

Results

```python
results = soup.find_all('div',{'class':'vehicle-card'})
```

```python
len(results)
```

```python
results[0]
```

Targeting the necessary data

```python
# Name
# Mileage
# Price
```

Name

```python
results[0].find('h2').get_text()
```

Mileage

```python
results[0].find('div',{'class': 'mileage'}).get_text()
```

Price

```python
results[0].find('span',{'class': 'primary-price'}).get_text()
```

Getting info of all results on the page

Putting everything in a list with a for-loop

```python
name = []
mileage = []
price = []

for result in results:

    #name
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append('n/a')
    
     #mileage
    try:
        mileage.append(result.find('div',{'class': 'mileage'}).get_text())
    except:
        mileage.append('n/a')
    
    #price
    try:
        price.append(result.find('span',{'class': 'primary-price'}).get_text())
    except:
        price.append('n/a')
    
```

```python
len(price)
```

Creating Pandas Dataframe

```python
car_dealer = pd.DataFrame({'Name':name, 'Mileage':mileage, 'Price':price})
```

```python
car_dealer
```

Output in Excel

```python
car_dealer.to_excel('single_page_car.xlsx',index=False)
```

## Pagination

```python
name = []
mileage = []
price = []



for i in range(1,501):
    
    #website in a variable
    website = 'https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&list_price_max=40000&makes[]=&maximum_distance=30&models[]=&stock_type=used&zip=75010'
    
    #request to website
    response = requests.get(website)
    
    #soup object
    soup = BeautifulSoup(response.content,'html.parser')
    
    #results
    results = soup.find_all('div',{'class':'vehicle-card'})
    
    
    for result in results:

        #name
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')

         #mileage
        try:
            mileage.append(result.find('div',{'class': 'mileage'}).get_text())
        except:
            mileage.append('n/a')

        #price
        try:
            price.append(result.find('span',{'class': 'primary-price'}).get_text())
        except:
            price.append('n/a')
```

```python
car_dealer = pd.DataFrame({'Name':name, 'Mileage':mileage, 'Price':price})
```

```python
car_dealer
```

Output to Excel

```python
car_dealer.to_excel('multi_page_car.xlsx',index=False)
```
